```python
import joblib
import numpy as np

tracking_results = joblib.load(
    "tracking_results.pth"
)

print("TYPE:")
print(type(tracking_results))

print("\nTOP LEVEL KEYS:")
print(tracking_results.keys())

for person_id, person_data in tracking_results.items():

    print("\n========================")
    print("PERSON ID:", person_id)
    print("========================")

    print("TYPE:", type(person_data))

    if isinstance(person_data, dict):

        print("\nFIELDS:")
        for k, v in person_data.items():

            print(f"\n{k}")

            print("type:", type(v))

            if isinstance(v, np.ndarray):
                print("shape:", v.shape)
                print("dtype:", v.dtype)

                # show small sample
                flat = v.reshape(-1)
                print("sample:", flat[:10])

            else:
                print(v)

    break
```
